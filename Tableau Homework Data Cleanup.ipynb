```python
import pandas as pd
```

```python
list_of_months = [201601,201602,201603,201604,201605,201606,201607,201608,201609,201610,201611,201612,201701,201702,201703,201704,201705,201706,201707,201708,201709,201710,201711,201712,201801,201802,201803,201804,201805,201806,201807,201808,201809,201810,201811,201812,201901,201902,201903,201904,201905]
list_of_columns = []
li = []
for month in list_of_months: 
    #print(month)
    filepath = ("data/JC-" + str(month) +"-citibike-tripdata.csv")
    df = pd.read_csv(filepath, index_col=None, header=0)
    month = str(month)
    df['timeperiod'] = month
    df['year'] = month[:4]
    df['month'] = month[-2:]
    
    for col in df.columns: 
        df = df.rename(columns={col: col.replace(" ", "").lower()})
        list_of_columns.append(col.replace(" ", "").lower())
        
        #print(str(month) + col)
    
    li.append(df)

def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    for x in unique_list: 
        print(x)
        
        
unique(list_of_columns)
#print(list_of_columns)
frame = pd.concat(li, axis=0, ignore_index=True)

frame.head(50)

frame.count()
```
